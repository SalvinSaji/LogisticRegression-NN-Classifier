```python
df.drop(columns={'PassengerId','Name', 'Ticket', 'Cabin', 'Embarked'},inplace = True)
df
```

```python
df['Age'].isna().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean()).round(2)
df
```

```python
pd.get_dummies(df['Sex'])
```

```python
pd.get_dummies(df['Pclass'])
```

```python
df['male'] = pd.get_dummies(df['Sex'])['male']
df['2'] = pd.get_dummies(df['Pclass'])[2]
df['3'] = pd.get_dummies(df['Pclass'])[3]
df
```

```python
df.drop({'Pclass','Sex'},axis = "columns",inplace = True)
df
```

```python
df.rename({'male':'sex','2':'pclass_2','3':'pclass_3'},axis = 'columns',inplace = True)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,plot_confusion_matrix
from sklearn.linear_model import LogisticRegression
```

```python
X = df.drop('Survived',axis=1)
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state = 42)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
lr_model= LogisticRegression()
lr_model.fit(X_train,y_train)
y_pred=lr_model.predict(X_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
accuracy_score(y_test,y_pred)
```

```python
plot_confusion_matrix(lr_model, X_test, y_test)
plt.title("Confusion Matrix (Logistic Regression)")
plt.show()
```

```python
mlpc = MLPClassifier(hidden_layer_sizes = (11,11,11),max_iter = 400)
mlpc.fit(X_train,y_train)
pred_mlpc = mlpc.predict(X_test)
print(classification_report(y_test,pred_mlpc))
print(confusion_matrix(y_test,pred_mlpc))
```

```python
accuracy_score(y_test,pred_mlpc)
```

```python
plot_confusion_matrix(mlpc, X_test, y_test)
plt.title("Confusion Matrix (Neural Network)")
plt.show()
```
